```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
tables = pd.read_html("https://en.wikipedia.org/wiki/List_of_countries_by_carbon_dioxide_emissions")
countries = tables[1]
countries
```

```python
#rename column names
countries.columns = ['country', '1990', '2005', '2017', '2021', 'emiss2017%world', 'emiss20171990%change', 'emiss2017perlandarea', 'emiss2017percapita','emiss2018incllucf', '2018excllucf']
countries
```

```python
#filter out rows that aren't countries
countries = countries.drop(labels=[0, 1, 2, 67], axis=0) # drop rows 0, 1 and 2 (world) and row 67 (Europe)
countries
```

```python
# make a graph of the top 5 biggest CO2 producers in the world
# x-axis 1990, 2005, 2017, 2021
# y-axis fossile CO2 emmissions (in Mt)
# 5 biggest CO2 producers in the world (based on the latest measurement = 2021)

countries['2021'] = pd.to_numeric(countries['2021'], downcast='float') #sort numeric values in column '2021' (type on wikipedia is different from other 'year'columns)
countries.sort_values(by=['2021'], ascending=False).head(5) # select top 5 rows from the sorted dataframe
```

```python
# make a graph of the top 5 biggest CO2 producers in the world
# x-axis 1990, 2005, 2017, 2021
# y-axis fossile CO2 emmissions (in Mt)
# top 5 biggest CO2 producers in the world (based on the latest measurement = 2021) > see data in table above

fig, ax = plt.subplots(figsize=(12, 8))
x = [1990, 2005, 2017, 2021]

China = [2397.048, 6263.064, 10877.218, 12466.320] #see values in table above 
United_states = [5085.897, 5971.571, 5107.393, 4752.080] #see values in table above
India = [605.968, 1210.754, 2454.774, 2648.780] #see values in table above 
Russia = [2378.921, 1733.950, 1764.866, 1942.540] #see values in table above
Japan = [1149.400, 1276.863, 1320.776, 1084.690] #see values in table above
#unsolved => numbers in column '2021' exactly similar to numbers in columns '1990', '2005', '2017'

w = 0.1 # width of bar

bar1 = np.arange(len(x))
bar2 = [i+w for i in bar1] #bars next to each other
bar3 = [i+w for i in bar2] #bars next to each other
bar4 = [i+w for i in bar3] #bars next to each other
bar5 = [i+w for i in bar4] #bars next to each other

ax.bar(bar1, China, w, label="China", color='b')
ax.bar(bar2, United_states, w, label="United States", color='g')
ax.bar(bar3, India, w, label="India", color='r')
ax.bar(bar4, Russia, w, label="Russia", color='c')
ax.bar(bar5, Japan, w, label="Japan", color='m')

ax.set_title("Top 5 biggest CO2 producers in the world")
ax.set_xlabel("Year")
ax.set_ylabel("fossile CO2 Emission (in Mt)")
ax.set_xticks(bar1+2*w,x); #set year every time in the middle of the 5 bars
ax.grid(axis = 'y') #set only grid horizontally
ax.legend(); #plot legend
ax.plot()
```

```python
#graph 2 worst and best changers
#we calculate the relative change and sort by that change
#we'll get a list with countries that are changing for the better at one end and countries that are changing less well on the other end.
#find the top 3 countries with regards to lowering these emissions and plot their results.
countries_abovefivemt = countries.loc[countries['2021'] > 5] #select emission (values) larger than 5 Mton
countries_abovefivemt.sort_values(by='emiss20171990%change', ascending=False).head(4) # select top 4 rows from the sorted dataframe
countries_abovefivemt = countries_abovefivemt.drop(labels=[30], axis=0) # drop row 30 which has a values below 5 Mton in years before 2021
countries_abovefivemt.sort_values(by='emiss20171990%change', ascending=False).head(3) # select top 3 rows from the sorted dataframe
```

```python
#find the bottom 3 countries with regards to lowering these emissions and plot their results.
countries_abovefivemt['emiss20171990%change'] = countries_abovefivemt['emiss20171990%change'].str.replace('.', '', regex=True) #remove dots between numbers
countries_abovefivemt['emiss20171990%change'] = countries_abovefivemt['emiss20171990%change'].str.replace('%', '', regex=True) #remove % after numbers
countries_abovefivemt['emiss20171990%change'] = pd.to_numeric(countries_abovefivemt['emiss20171990%change'], downcast='signed') #change numbers into integers
countries_bottom = countries_abovefivemt.sort_values(by=['emiss20171990%change'], ascending=False).tail(~100) #select top 100 negative values > bottom values with the largest relative change relating to 1990 (=100%) = biggest reduction of CO2 emission compared to the CO2 emission in 1990
countries_bottom[::-1].head(4) #filter top 4 of the previous selection
countries_bottom[::-1].drop(labels = [10], axis = 0).head(3) # remove row 10 because it has less than 5 Mton emission in years before 2021
#note > unsolved problem in column 'emiss20171990%change' is that the numbers are 10 times bigger than should be, so -741% should be -74.1%
```
