```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
#%matplotlib inline
%reload_ext google.colab.data_table

df_gdp = pd.read_csv("/content/drive/MyDrive/gross-domestic-product.csv", encoding="ISO-8859-1")
df_gdp
```

```python
df_gdp2019 = df_gdp.loc[df_gdp['Year'] == 2019] # select year 2019, year of the graph in the source given
df_gdp2019
```

```python
df_gini = pd.read_csv("/content/drive/MyDrive/economic-inequality-gini-index.csv", encoding="ISO-8859-1")
df_gini
```

```python
df_gini2019 = df_gini.loc[df_gini['Year'] == 2019] # select year 2019, year of the graph in the source given
df_gini2019
```

```python
merged_data = df_gdp2019.merge(df_gini2019,on=["Entity"]) # merge two csv files with data of the year 2019 on column Entity
merged_data
```

```python
merged_data.drop(columns=['Code_x', 'Year_x', 'Code_y', 'Year_y'], inplace=True)
```

```python
merged_data.rename(columns={"Entity": "Country", "GDP (constant 2015 US$)": "GDP_country", "Gini coefficient": "Gini"}, inplace=True)
merged_data
```

```python
plt.figure(figsize=(15,15)) # difficult to find right plotsize because almost all values are very close to each other

sns.scatterplot(data=merged_data,x='GDP_country',y='Gini') # scatterplot of GDP per country versus Gini coefficient
for i in range(merged_data.shape[0]):
    plt.text(x=merged_data.GDP_country[i]+100000000000,y=merged_data.Gini[i],s=merged_data.Country[i], 
             fontdict=dict(color='red',size=8))

plt.xlim(merged_data.GDP_country.min(),merged_data.GDP_country.max()+5000000000000) # set x limit
plt.ylim(merged_data.Gini.min()-0.01,merged_data.Gini.max()+0.01) # set y limit

plt.title("GDP per country versus Gini coefficient (2020)") # title
plt.grid() # gridlines
plt.xlabel("GDP per country (constant 2015 US $)") # x label
plt.ylabel("Gini coefficient (index for degree of inequality in distribution of income/wealth)") # y label
plt.show()
```
